# Solution of Exercise

```python
def calc(a, b, op):
    if op == "+":
        return [a+b, "sum"]
    elif op == "-":
        return [a-b, "difference"]
    elif op == "*":
        return [a*b, "product"]
    elif op == "/":
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero")
        else:
            return [a/b, "division"]
        
x = 7
y = 0
t = calc(x, y, "*")
print("The ", t[1], " of ", x, " and ", y, " is ", t[0])
```
